```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Take only ONE feature (as required in experiment)
X = df['sepal length (cm)'].values

# Target (you can choose any feature as output)
y = df['petal length (cm)'].values

# Scatter plot (before training)
plt.scatter(X, y)
plt.title("Iris Data (Before Training)")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
m = len(X)

theta0 = 0   # intercept
theta1 = 0   # slope

learning_rate = 0.01
iterations = 1000

cost_history = []
```

```python
def hypothesis(x, theta0, theta1):
    return theta0 + theta1 * x
```

```python
def compute_cost(X, y, theta0, theta1):
    m = len(X)
    cost = 0

    for i in range(m):
        cost += (hypothesis(X[i], theta0, theta1) - y[i]) ** 2

    return cost / (2 * m)
```

```python
for _ in range(iterations):
    temp0 = 0
    temp1 = 0

    for i in range(m):
        error = hypothesis(X[i], theta0, theta1) - y[i]
        temp0 += error
        temp1 += error * X[i]

    theta0 = theta0 - (learning_rate / m) * temp0
    theta1 = theta1 - (learning_rate / m) * temp1

    cost_history.append(compute_cost(X, y, theta0, theta1))
```

```python
y_pred = theta0 + theta1 * X

plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, color='red', label="Best Fit Line")
plt.title("Best Fit Line (Iris Dataset)")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend()
plt.show()
```

```python
plt.plot(range(iterations), cost_history)
plt.title("Cost Function vs Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```
