```python
Sentence = """Myself Tanvi Kakadiya. I am 19 year old girl. Currently i am studying in marwadi university. I like to play badminton. My hometown is botad."""
```

```python
#importing basic libraries
import nltk
import networkx as nx
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python
#download requirements
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# from ast import keyword
def textrank_keyword(text,topn=5,window_size=4):
  words = word_tokenize(text.lower())
  stop_words = set(stopwords.words('english'))
  words = [w for w in words if w not in stop_words and w not in string.punctuation]
  graph = nx.Graph()

  for i ,word in enumerate(words):
    for j in range(i+1,window_size+i):
      if(j<len(words)):
        graph.add_edge(word,words[j])

  scores = nx.pagerank(graph)
  ranked_words = sorted(scores.items(),key=lambda x:x[1],reverse=True)
  keywords = [word for word, score in ranked_words[:topn]]
  return keywords
```

```python
print(textrank_keyword(Sentence))
```
